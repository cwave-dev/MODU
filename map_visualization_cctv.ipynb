```python
df = pd.read_csv('/content/drive/MyDrive/데이터포럼/Toilet_Data/서울교통공사_역사공중화장실정보_cctv.csv', encoding='cp949')
print(df.head(80))
```

```python
df = df.fillna(value="무")
cctv = []
nocctv = []

for i in range(len(df['x'])):
    tmp = []
    if (df['cctv'][i] == "유"):
        tmp.append(df['x'][i])
        tmp.append(df['y'][i])
        cctv.append(tmp)
        tmp = []
    if (df['cctv'][i] == "무"):
        tmp.append(df['x'][i])
        tmp.append(df['y'][i])
        nocctv.append(tmp)
    else:
      pass
nocctv
```

```python
from pandas.core.dtypes.cast import maybe_cast_to_datetime
map_osm = folium.Map(location=[37.56, 126.97],zoom_start=11)

for i in range(len(cctv)):
     folium.Marker([float(cctv[i][0]),float(cctv[i][1])], popup='CCTV 존재', icon=folium.Icon(color='pink', icon='info-sign')).add_to(map_osm)

map_osm
```

```python
from pandas.core.dtypes.cast import maybe_cast_to_datetime
map_osm = folium.Map(location=[37.56, 126.97],zoom_start=11)

for i in range(len(nocctv)):
     folium.Marker([float(nocctv[i][0]),float(nocctv[i][1])], popup='CCTV 존재 X', icon=folium.Icon(color='blue', icon='info-sign')).add_to(map_osm)

map_osm
```
