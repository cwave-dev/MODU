```python
df = pd.read_csv('/content/drive/MyDrive/데이터포럼/Toilet_Data/서울교통공사_역사공중화장실정보_child.csv', encoding='cp949')
print(df.head(5))
```

```python
mb = []
ms = []
m = []
wb = []

for i in range(len(df['mb'])):
    tmp = []
    if (df['mb'][i] > 0):
        tmp.append(df['x'][i])
        tmp.append(df['y'][i])
        mb.append(tmp)
        tmp = []
    if (df['ms'][i] > 0):
        tmp.append(df['x'][i])
        tmp.append(df['y'][i])
        ms.append(tmp)
        tmp = []
    if (df['wb'][i] > 0):
        tmp.append(df['x'][i])
        tmp.append(df['y'][i])
        wb.append(tmp)
        tmp = []
    if ((df['mb'][i] > 0) & (df['ms'][i] > 0)):
        tmp.append(df['x'][i])
        tmp.append(df['y'][i])
        m.append(tmp)
    else:
      pass


len(m)
# 값: 139 84 150 60
```

```python
from pandas.core.dtypes.cast import maybe_cast_to_datetime
map_osm = folium.Map(location=[37.56, 126.97],zoom_start=11)

for i in range(len(mb)):
     folium.Marker([float(mb[i][0]),float(mb[i][1])], popup='남자아이대변기', icon=folium.Icon(color='blue', icon='info-sign')).add_to(map_osm)

map_osm
```

```python
from pandas.core.dtypes.cast import maybe_cast_to_datetime
map_osm = folium.Map(location=[37.56, 126.97],zoom_start=11)

for i in range(len(ms)):
     folium.Marker([float(ms[i][0]),float(ms[i][1])], popup='남자아이소변기', icon=folium.Icon(color='purple', icon='info-sign')).add_to(map_osm)

map_osm
```

```python
from pandas.core.dtypes.cast import maybe_cast_to_datetime
map_osm = folium.Map(location=[37.56, 126.97],zoom_start=11)

for i in range(len(m)):
     folium.Marker([float(m[i][0]),float(m[i][1])], popup='남자아이변기', icon=folium.Icon(color='darkblue', icon='info-sign')).add_to(map_osm)

map_osm
```
